```python
import pandas as pd
```

```python
data = pd.read_csv("./mnist_train.csv",header=None,names = ["label_"+str(i) for i in range(785)])
```

```python
data.head()
```

```python
label = data["label_0"]
```

```python
label
```

```python
feature = data[["label_"+str(i) for i in range(1,785)]]
```

```python
feature
```

```python
import numpy as np
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB()
```

```python
clf.fit(feature, label)
```

```python
data_test = pd.read_csv("./mnist_test.csv",header=None,names = ["label_"+str(i) for i in range(785)])
test_feature = data_test[["label_"+str(i) for i in range(1,785)]]
test_label = data_test["label_0"]
```

```python
data_test
```

```python
test_p = clf.predict(test_feature)
```

```python
sum(test_label==test_p)/data_test.shape[0]
```
